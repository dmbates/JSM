# Fitting mixed models in lme4 and MixedModels

The [`lme4`](https://github.com/lme4/lme4) package for [`R`](http://R-project.org) and the [`MixedModels`](https://github.com/dmbates/MixedModels.jl) package for [`Julia`](http://JuliaLang.org) provide functions for fitting and analyzing mixed-effects models.  The algorithms used are similar but not exactly the same.

49 of the model/dataset combinations were fit with `lmm` using four different optimizers from the [`NLopt`](https://github.com/stevengj/NLopt.jl) package for Julia and using the built-in optimizers of `lmer` and optimizers from the `nloptr` (the underlying optimization code is the same as that in `NLopt`) and `optimx` packages for `R`.  The largest dataset, `ml10m`, with over 10 million observations, was fit using `LN_BOBYQA` in `Julia` and `NLOPT_LN_BOBYQA` in `R`.  These were generally the fastest and most reliable of the optimizers considered.

```julia
using DataFrames, DataFramesMeta, RCall
timings = @linq readtable("results50.csv", makefactors = true) |>
    transform(lang = pool(map(x -> startswith(x, "LN_") ? "J" : "R", :optimizer)),
        model = pool(map(x -> split(x, ':')[2], :dsmodel)),
        ds = pool(map(x -> split(x, ':')[1], :dsmodel))) |>
    transform(dsmodel = pool(map((d,m) -> string(d, ':', Int(m)), :ds, :model.refs)))
excess = @by(timings, :dsmodel, optimizer = :optimizer, 
    excess = :objective .- minimum(:objective))
timings = join(timings, excess, on = [:dsmodel,:optimizer]);
head(timings)
```

These columns are:
- `dsname`: a string constructed from the dataset name and the model specification
- `n`: the number of observations in the data set
- `nopt`: the dimension of the vector $\theta$ over which the optimization takes place
- `p`: the dimension of the fixed-effects parameter vector, $\beta$
- `optimizer`: the name of the optimizer used
- `objective`: the minimum value of the objective achieved
- `elapsed`: elapsed time (seconds)
- `neval`: number of evaluations until convergence is declared
- `excess`: difference between the achieved objective value and the minimum achieved by any optimizer

## Comparing the accuracy of the optimizers

For optimization problems such as considered here we cannot "prove" convergence.  The best we can do is to use several different optimizers on the same problem and assume that the minimum achieved by any of the optimizers is close to the actual minimum.  We then must determine if the results of the other optimizers are "close enough" to this value.  For this we calculate `:excess`, the difference between the achieved objective value and the minimum over the all the optimizers applied to that `dsmodel`.

This objective is negative twice the log-likelihood, the quantity used in a likelihood ratio test comparing nested models.  This leads us to define "close enough" as within 0.05 of the minimum over all the optimizers.

```julia
@by(timings, :optimizer, nmod = length(:excess), nfail = sum(:excess .> 0.05), 
    maxexcess = maximum(:excess))
```

# Comparing speed

To compare the elapsed times we consider only those cases which are considered to have converged, i.e. `excess < 0.05`, and calculate the ratio of the elapsed time to the minimum elapsed time.

```julia
cvgd = @where(timings, :excess .< 0.05)
ratio = @by(cvgd, :dsmodel, optimizer = :optimizer,
    tratio = :elapsed ./ minimum(:elapsed))
cvgd = join(cvgd, ratio, on = [:dsmodel, :optimizer])
```

```julia
@by(cvgd, :optimizer, min = minimum(:tratio), gmean = geomean(:tratio), 
   max = maximum(:tratio))
```
